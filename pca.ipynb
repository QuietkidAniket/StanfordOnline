```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
X = np.array([[1,1], [2,1], [3,2], [-1,-1], [-2,-1], [-3,-2]])
X
```

```python
plt.plot(X[:,0] , X[:,1], 'ro')
```

```python
X.shape
```

```python
pca_1 = PCA(n_components=1); pca_1.fit(X)
pca_1.explained_variance_ratio_
```

```python
X_trans_1 = pca_1.transform(X)
X_reduced_1 = pca_1.inverse_transform(X_trans_1)
```

```python
# PCA projection
X_trans_1
```

```python
X_reduced_1
```

```python
plt.plot(X[:, 0], X[:, 1], 'ro', label =  "original datapoints")
plt.plot(X_reduced_1[:,0], X_reduced_1[:,1], 'bx', label = "PCA decomposition")
plt.plot(X_reduced_1[:, 0], X_reduced_1[:, 1], 'g--', label = "Principle Component")
plt.legend()
plt.show()
```

```python
pca_2 = PCA(n_components=2); pca_2.fit(X)
pca_2.explained_variance_ratio_
```

```python
X_trans_2 = pca_2.transform(X)
X_reduced_2 = pca_2.inverse_transform(X_trans_2)
```

```python
X_trans_2
```

```python
X_reduced_2
```

```python
plt.plot(X[:, 0], X[:, 1], 'ro', label =  "original datapoints")
plt.plot(X_reduced_2[:,0], X_reduced_2[:,1], 'bx', label = "PCA decomposition")

plt.legend()
plt.show()
```
